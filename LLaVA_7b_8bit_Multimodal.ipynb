# Step 2: Install Libraries
The use of threading in Python to run a subprocess concurrently. It launches a Python module 'llava.serve.controller' as a server using the 'subprocess' module and runs it in a separate thread.

```python
import threading
import subprocess
threading.Thread(target=lambda: subprocess.run(['python3', '-m', 'llava.serve.controller', '--host', '0.0.0.0', '--port', '10000'], check=True), daemon=True).start()
```

A model worker server is initiated in the background with specified parameters such as host address, controller URL, port number, worker URL, model path, and the option to load an 8-bit model. The threading approach allows the main program to continue execution while the server operates independently.

```python
import threading
import subprocess
command = [
    'python3', '-m', 'llava.serve.model_worker',
    '--host', '0.0.0.0',
    '--controller', 'http://localhost:10000',
    '--port', '40000',
    '--worker', 'http://localhost:40000',
    '--model-path', '4bit/llava-v1.5-7b-5GB',
    '--load-8bit'
]
threading.Thread(target=lambda: subprocess.run(command, check=True, shell=False), daemon=True).start()
```
